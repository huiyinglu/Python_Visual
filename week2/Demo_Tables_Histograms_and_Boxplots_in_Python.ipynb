### Visualizing Data in Python
#### Tables, Histograms, Boxplots, and Slicing for Statistics

When working with a new dataset, one of the most useful things to do is to begin to visualize the data. By using tables, histograms, box plots, and other visual tools, we can get a better idea of what the data may be trying to tell us, and we can gain insights into the data that we may have not discovered otherwise.

Today, we will be going over how to perform some basic visualisations in Python, and, most importantly, we will learn how to begin exploring data from a graphical perspective.

```python
# We first need to import the packages that we will be using
import seaborn as sns # For plotting
import matplotlib.pyplot as plt # For showing plots

# Load in the data set
tips_data = sns.load_dataset("tips")
```

#### Visualizing the Data - Tables
When you begin working with a new data set,  it is often best to print out the first few rows before you begin other analysis. This will show you what kind of data is in the dataset, what data types you are working with, and will serve as a reference for the other plots that we are about to make. 

```python
# Print out the first few rows of the data
tips_data.head()
```

#### Describing Data
Summary statistics, which include things like the mean, min, and max of the data, can be useful to get a feel for how large some of the variables are and what variables may be the most important. 

```python
# Print out the summary statistics for the quantitative variables
tips_data.describe()
```

#### Creating a Histogram

After we have a general 'feel' for the data, it is often good to get a feel for the shape of the distribution of the data.

```python
# Plot a histogram of the total bill
sns.distplot(tips_data["total_bill"], kde = False).set_title("Histogram of Total Bill")
plt.show()
```

```python
# KRISTIN: use displot to remove the warning
# Note: We need to use numerical variables for histgogram 
# Note: sns.xxxplot().set_title will sometimes generate errors. Use sns.xxxplot().set instead
#sns.displot(tips_data["total_bill"], kde=False).set_title("Histogram of Total Bill")
sns.displot(tips_data["total_bill"], kde=False).set(title = "Histogram of Total Bill")
plt.show()
```

```python
# KRISTIN: Or, use histplot to remove the warning
#sns.histplot(tips_data["total_bill"], kde=False).set_titles("Histogram of Total Bill")

sns.histplot(tips_data["total_bill"], kde=False).set(title = "Histgram of Total Bill");
plt.show()
```

```python
# KRISTIN: another way is pass tips_data as data parameter and pass "total_bill" as x parameter.
sns.histplot(data = tips_data, x = "total_bill").set(title = "Histgram of Total Bill")
plt.show()
```

```python
# Plot a histogram of the Tips only
sns.distplot(tips_data["tip"], kde = False).set_title("Histogram of Total Tip")
plt.show()
```

```python
# KRISTIN: use displot to remove warning
sns.displot(tips_data["tip"], kde = False).set(title = "Histogram of Total Tip")
plt.show()
```

```python
# KRISTIN: use hisplot to remove warning
sns.histplot(tips_data["tip"], kde = False).set(title = "Histogram of Total Tip")
plt.show()
```

```python
# Plot a histogram of both the total bill and the tips'
sns.distplot(tips_data["total_bill"], kde = False)
sns.distplot(tips_data["tip"], kde = False).set_title("Histogram of Both Tip Size and Total Bill")
plt.show()
```

```python
# KRISTIN: use displot to remove warning. However, we got 2 graphs.
sns.displot(tips_data["total_bill"], kde = False)
sns.displot(tips_data["tip"], kde = False).set(title = "Histogram of Both Tip Size and Total Bill")
plt.show()
```

```python
# KRISTIN: use histplot to remove warning. Also add color parameter.
sns.histplot(tips_data["total_bill"], kde = False)
sns.histplot(tips_data["tip"], kde = False, color = 'orange').set(title = "Histogram of Both Tip Size and Total Bill")
plt.show()
```

#### Creating a Boxplot

Boxplots do not show the shape of the distribution, but they can give us a better idea about the center and spread of the distribution as well as any potential outliers that may exist. Boxplots and Histograms often complement each other and help an analyst get more information about the data

```python
# Create a boxplot of the total bill amounts
sns.boxplot(tips_data["total_bill"]).set_title("Box plot of the Total Bill")
plt.show()
```

```python
# KRISTIN: Pass tips_data as a keyword arg data; Pass "total_bill" as a keyword arg x to remove the warning
# Note: we need to use a numerical variable for boxplot
# Create a boxplot of the total bill amounts
sns.boxplot(data = tips_data, x = "total_bill").set(title = "Box plot of the Total Bill")
plt.show()
```

```python
# Create a boxplot of the tips amounts
sns.boxplot(tips_data["tip"]).set_title("Box plot of the Tip")
plt.show()
```

```python
# KRISTIN: pass data and y.
sns.boxplot(data = tips_data, y = "tip").set(title = "Box plot of the Tip")
plt.show()
```

```python
# Create a boxplot of the tips and total bill amounts - do not do it like this
sns.boxplot(tips_data["total_bill"])
sns.boxplot(tips_data["tip"]).set_title("Box plot of the Total Bill and Tips")
plt.show()
```

```python
# KRISTIN: pass data and keyword arg x.
# Create a boxplot of the tips and total bill amounts - do not do it like this
sns.boxplot(data = tips_data, x = "total_bill")
sns.boxplot(data = tips_data, x = "tip").set(title = "Box plot of the Total Bill and Tips")
plt.show()
```

```python
# KRISTIN: scatterplot is commonly used for checking the linear relationship between 2 numberial variables.
sns.scatterplot(data = tips_data, x = "total_bill", y = "tip")
plt.show()
```

```python
# KRISTIN: regplot is commonly used for checking the linear relationship between 2 numberial variables.
sns.regplot(data = tips_data, x = "total_bill", y = "tip")
plt.show()
```

#### Creating Histograms and Boxplots Plotted by Groups

While looking at a single variable is interesting, it is often useful to see how a variable changes in response to another. Using graphs, we can see if there is a difference between the tipping amounts of smokers vs. non-smokers, if tipping varies according to the time of the day, or we can explore other trends in the data as well.

```python
# Create a boxplot and histogram of the tips grouped by smoking status
sns.boxplot(x = tips_data["tip"], y = tips_data["smoker"])
plt.show()
```

```python
# KRISTIN: We have two variables now. One is numerical, the other one is categorical. 
# Use x for the categorical variable, and y for numerical variable is fine.
sns.boxplot(data = tips_data, x = "day", y = "tip")
plt.show()
```

```python
# KRISTIN: How about three variables? One is numerical, the other two are categorical. 
# Use x for the major categorical variable, y for numerical variable, and hue for another categorical variable.
sns.boxplot(data = tips_data, x = "day", y = "tip", hue='sex')
plt.show()
```

```python
# Create a boxplot and histogram of the tips grouped by time of day
sns.boxplot(x = tips_data["tip"], y = tips_data["time"])

g = sns.FacetGrid(tips_data, row = "time")
g = g.map(plt.hist, "tip")
plt.show()
```

```python
# Create a boxplot and histogram of the tips grouped by the day
sns.boxplot(x = tips_data["tip"], y = tips_data["day"])

g = sns.FacetGrid(tips_data, row = "day")
g = g.map(plt.hist, "tip")
plt.show()
```

```python
#KRISTIN: use col, row and hue in FacetGrid for 3 categorical variables.
g = sns.FacetGrid(tips_data, col = 'day', row = "time", hue='sex')
g = g.map(plt.scatter, "total_bill", "tip")
g.add_legend()
plt.show()
```

```python
# KRISTIN: we may use seaborn PairGrid for two numerical variables.
# Note: We use histplots on diag, use scatterplot/regplot for offdiag.
g = sns.PairGrid(tips_data, vars=["total_bill", "tip"], hue="sex")
g.map_diag(sns.histplot)
#g.map_offdiag(sns.scatterplot)
g.map_offdiag(sns.regplot)
g.add_legend()
plt.show()
```
